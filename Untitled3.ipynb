```
import matplotlib.pyplot as plt
import matplotlib.image as im
import numpy as np

as2 = im.imread(os.getcwd()+'/NEU-CLS-64/pa/1.jpg')
plt.figure()
mini = as2.min()
maxi = as2.max()
as1 = (255)/(maxi - mini)*(as2-mini)
plt.figure()
plt.imshow(as2)
plt.show()
plt.imshow(as1)
plt.show()
```

```
bnr = np.zeros(as2.shape)
print(bnr.shape)
```

```
threshold = 100
count = 0
for i in range(0,as1.shape[0]):
  for j in range(0,as1.shape[1]):
    if(as1[i][j] >= threshold):
      bnr[i][j] = 1
      count+=1
   
plt.imshow(bnr)
print(count)
```
